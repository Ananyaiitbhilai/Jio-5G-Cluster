```python
df_jc1_interpolate1 = df_jc1_interpolate1[['count_of_Customers','coverage_percent', 'high_speed_percent ']]
print(df_jc1_interpolate1.shape)
```

```python
train_df=df_jc1_interpolate1[:-12]
test_df=df_jc1_interpolate1[-12:]
```

```python
print(test_df.shape)
```

```python
model = VAR(train_df.diff()[1:])
```

```python
sorted_order=model.select_order(maxlags=20)
print(sorted_order.summary())
```

```python
var_model = VARMAX(train_df, order=(3,0),enforce_stationarity= True)
fitted_model = var_model.fit(disp=False)
print(fitted_model.summary())
```

```python
n_forecast = 12
predict = fitted_model.get_prediction(start=len(train_df),end=len(train_df) + n_forecast-1)

predictions=predict.predicted_mean
```

```python
predictions.columns=['count_of_Customers_predi','coverage_percent_predi', 'high_speed_percent_pred ']
predictions
```

```python
test_vs_pred=pd.concat([test_df,predictions],axis=1)
```

```python
test_vs_pred.plot(figsize=(12,5))
```

```python
from sklearn.metrics import mean_squared_error
import math 
from statistics import mean

rmse_count_of_Customers=math.sqrt(mean_squared_error(predictions['count_of_Customers_predi'],test_df['count_of_Customers']))
print('Mean value of count of customers is : {}. Root Mean Squared Error is :{}'.format(mean(test_df['count_of_Customers']),rmse_count_of_Customers))

rmse_coverage_percent=math.sqrt(mean_squared_error(predictions['coverage_percent_predi'],test_df['coverage_percent']))
print('Mean value of coverage_percent is : {}. Root Mean Squared Error is :{}'.format(mean(test_df['coverage_percent']),rmse_coverage_percent))

rmse_high_speed_percent=math.sqrt(mean_squared_error(predictions['high_speed_percent_pred '],test_df['high_speed_percent ']))
print('Mean value of high_speed_percent is : {}. Root Mean Squared Error is :{}'.format(mean(test_df['high_speed_percent ']),rmse_high_speed_percent))
```
